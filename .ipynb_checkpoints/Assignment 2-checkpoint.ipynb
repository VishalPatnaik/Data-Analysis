```python
def make_random_indices(seed = None, training_fraction = 0.75):
    if seed :
        np.random.seed(seed = seed)
    n = len(v_levels)
    n_train = int(n * 3.0 / 4)

    i_shuffle = np.random.permutation(n)        
    i_train = i_shuffle[ : n_train]               
    i_test = i_shuffle[n_train : ]                
    assert set(i_train) & set(i_test) == set([])
    return i_train, i_test
```

The above function gives the random indices taking 3/4th to train and 1/4th to test.

```python
import itertools
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import confusion_matrix, classification_report



def performance_report(y_test, y_pred):
    plot_confusion_matrix(y_test, y_pred)
    classification_error_rate = 1.0 * sum(1 for (x, y) in zip(y_test, y_pred) if x != y) / len(list(y_test))
    
    print("Classification Error Rate")
    print(classification_error_rate)
    
    print("\nClassification Report")
    return classification_report(y_test, y_pred)
    

def plot_confusion_matrix(y_test, y_pred,
                          title = 'Confusion matrix',
                          cmap = plt.cm.Blues, normalize=True):
   
    cm = confusion_matrix(y_test, y_pred)
    np.set_printoptions(precision = 2)

    plt.figure()
    fig, ax = plt.subplots(figsize = (9, 7.2))
    plt.rcParams.update({'font.size': 15})
    plt.imshow(cm, interpolation = 'nearest', cmap = cmap)
    plt.title(title)
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation = 45)
    plt.yticks(tick_marks, classes)
    cm = cm.astype('float') / cm.sum(axis = 1)[:, np.newaxis]
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, "{0:.3f}".format(cm[i, j]),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')

def feature_importance_report_tree(clf, features):
    feature_importances = zip(features, clf.feature_importances_)
    rows = []
    for f, imp in sorted(feature_importances, key = lambda x : x[1], reverse = True):
        rows.append([f, imp])
    v_rep = pd.DataFrame(rows, columns = ['Feature', 'Relative_Importance'])
    return v_rep
```
