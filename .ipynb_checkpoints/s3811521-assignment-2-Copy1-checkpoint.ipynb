```python
import matplotlib.pyplot as plt
import seaborn as sns

class PlotManager(object):
    def __init__(self):
        self.i = 1
    def save_last_plt(self):
        plt.savefig('Figure_{}.png'.format(self.i))
        self.i += 1

plt_manager = PlotManager()
```

```python
def plot_expression_values():
    for (i, var_sub) in enumerate(var_v_sub):
        v_s = melt_v.loc[melt_v.variable.isin(var_sub)]
        sns.violinplot(x = "variable", y = "value", hue = "variable", width = 1, data = v_s, scale = "area", orient = "v")     
        sns.despine(left = True)
        plt_manager.save_last_plt()
        yield plt.show()
```

```python
to_ignore = ['MouseNumber','MouseVersion']
class_labels = ['Genotype','Treatment','Behavior','class']
# the data frame which consists only of the target variables
v_class = v_mice[class_labels] 
# the data frame which consists only of the protein expression levels
v_levels = v_mice.drop(class_labels + to_ignore, axis = 1)
v_levels
```

```python
from sklearn import preprocessing
```

```python
scaler = preprocessing.StandardScaler()
v_levels[:] = scaler.fit_transform(v_levels[:])
v_levels.head()
```

```python
normalizer = preprocessing.Normalizer()
v_levels[:] = normalizer.fit_transform(v_levels[:])
v_levels
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
v_levels[:] = min_max_scaler.fit_transform(v_levels[:])
v_levels
```

```python
def make_random_indices(seed = None, training_fraction = 0.67):
    if seed :
        np.random.seed(seed = seed)
    n = len(v_levels)
    n_train = int(n * 2.0 / 3)

    i_shuffle = np.random.permutation(n)        # if n = 7, i_shuffle = [6 4 2 0 3 1 5]
    i_train = i_shuffle[ : n_train]               # then i_train = [6 4 2 0]
    i_test = i_shuffle[n_train : ]                # and i_test = [3,1,5]
    assert set(i_train) & set(i_test) == set([])
    return i_train, i_test
```

```python
import itertools
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import confusion_matrix, classification_report



def performance_report(y_test, y_pred, classes):
    plot_confusion_matrix(y_test, y_pred, classes)
    classification_error_rate = 1.0 * sum(1 for (x, y) in zip(y_test, y_pred) if x != y) / len(list(y_test))
    print(classification_error_rate)
    return classification_report(y_test, y_pred)
    

def plot_confusion_matrix(y_test, y_pred, classes,
                          title = 'Confusion matrix',
                          cmap = plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    # Compute confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    np.set_printoptions(precision = 2)


    # Plot normalized confusion matrix
    plt.figure()
    fig, ax = plt.subplots(figsize = (9, 7.2))
    plt.rcParams.update({'font.size': 15})
    plt.imshow(cm, interpolation = 'nearest', cmap = cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation = 45)
    plt.yticks(tick_marks, classes)
    cm = cm.astype('float') / cm.sum(axis = 1)[:, np.newaxis]
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, "{0:.3f}".format(cm[i, j]),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')

def feature_importance_report(tree, features):
    feature_importances = zip(features, tree.feature_importances_)
    rows = []
    for f, imp in sorted(feature_importances, key = lambda x : x[1], reverse = True):
        rows.append([f, imp])
    v_rep = pd.DataFrame(rows, columns = ['Feature', 'Relative_Importance'])
    return v_rep
```

```python
def run_classifier(clf, class_labels, seed = None):
    i_train, i_test = make_random_indices(seed = None)
    x_train = v_levels.iloc[i_train]
    y_train = v_class.iloc[i_train][class_labels]

    x_test = v_levels.iloc[i_test]
    y_test = v_class.iloc[i_test][class_labels]
    clf.fit(x_train, y_train)
    y_pred = tree.predict(x_test)

    class_names = set(v_class[class_labels].values)    

    print(performance_report(y_test, y_pred, class_names))
    print(clf.score(x_test, y_test))
    
    plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()
run_classifier(tree, 'class', seed = 1)
```

```python
v_imp = feature_importance_report(tree, v_levels.columns)
v_imp
```

```python
from sklearn.neighbors import KNeighborsClassifier  
knn = KNeighborsClassifier(5)
run_classifier(knn, 'class', seed = 1)
```
